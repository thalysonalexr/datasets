```
import matplotlib.pyplot as plt
```

```
# y = x**2 + 2 * x + 5
data_x = []
data_y = []

for x in range(1, 100):
  y = x**2 + (2 * x) + 5
  data_x.append(x)
  data_y.append(y)
  
plt.plot(data_x, data_y)
```

```
data_x = []
data_y = []

for x in range(-100, 100):
  y = x**2 + (2 * x) + 5
  data_x.append(x)
  data_y.append(y)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Parábola')
plt.plot(data_x, data_y)
```
