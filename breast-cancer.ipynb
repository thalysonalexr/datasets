```python
import sklearn
```

```python

from sklearn.datasets import load_breast_cancer

# Carregar o dataset
data = load_breast_cancer()

data
```

```python

# Organizar nossos dados
label_names = data['target_names']
labels = data['target']
feature_names = data['feature_names']
features = data['data']
```

```python

# Olhando para os nossos dados
print(label_names)
print(labels[0])
print(feature_names[0])
print(features[0])
```

```python

from sklearn.model_selection import train_test_split

# Dividir nossos dados
train, test, train_labels, test_labels = train_test_split(features,
                                                          labels,
                                                          test_size=0.15,
                                                          random_state=42)
```

```python

from sklearn.naive_bayes import GaussianNB

# Inicializar nosso classificador
gnb = GaussianNB()

# Treinar nosso classificador
model = gnb.fit(train, train_labels) 
```

```python


# Fazer previs�es
preds = gnb.predict(test)
print(preds)
```

```python

from sklearn.metrics import accuracy_score

# Avaliar a precis�o
print(accuracy_score(test_labels, preds))
```

```python
#arvore de decis�o
from sklearn import tree

tr = tree.DecisionTreeRegressor()

tr.fit(train,train_labels)

preds2 = tr.predict(test)

print(accuracy_score(test_labels, preds2))
```

```python
import graphviz 
dot_data = tree.export_graphviz(tr, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("breast") 
```
